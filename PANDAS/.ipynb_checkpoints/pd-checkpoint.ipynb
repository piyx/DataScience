# Making changes to data

```python
# Adding a new column total
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df
```

```python
# Remove column/columns
df = df.drop(columns=['Total'])
df
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df
```

```python
# Reorder columns (Moving 'Total' column)
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df
```

# Saving our data

```python
# Save as csv
df.to_csv('modified.csv', index=False) # Do not include the index column
```

```python
# Save as excel
df.to_excel('modified.xlsx', index=False)
```

```python
# Tab seperated file
df.to_csv('modified.txt', index=False, sep='\t')
```

# Filtering data

```python
df.loc[df['Type 1'] == 'Grass']
```

```python
# Using and
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
# Using or
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
# Reset index
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# Remove all rows that have 'Mega' in their name
df.loc[~df['Name'].str.contains('Mega')] # ~ stands for not
```

```python
# Using regex
import re

df.loc[df['Type 1'].str.contains('fire|grass', regex=True, flags=re.I)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', regex=True, flags=re.I)] # Name sarts with 'Pi'
```

# Conditional changes

```python
# Changing column values based on condition
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
# Change multiple parameters
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Z', True]
df
```

# Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')
df
```

```python
df.groupby(['Type 1']).mean()
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False) # Dragon OP
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
# The above output looks weird. Fixing it

df['Count'] = 1 # Making a new row and giving a count of 1 to every row
df
```

```python
df.groupby(['Type 1']).count()['Count']
```

```python
# Group by multiple columns
df.groupby(['Type 1', 'Type 2']).count()['Count']
```
